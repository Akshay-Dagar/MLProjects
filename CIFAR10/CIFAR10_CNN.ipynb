```python
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

# Feature Scaling

```python
#Getting Train and test data
x_train=(images_train.reshape(images_train.shape[0],-1))
x_test=(images_test.reshape(images_test.shape[0],-1))
y_test=labels_test
y_train=labels_train
```

```python
#Feature Scaling:
start=time.time()
sc=StandardScaler() 
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(time.time()-start)                       #since the dataset is huge time taken for execution is high for fit functions
```

```python
x_train=(x_train.reshape(images_train.shape))
x_test=(x_test.reshape(images_test.shape))
```

# Defining CNN Architecture:

```python
#intializing some constants:
input_width=x_train.shape[1]
input_height=x_train.shape[2]
input_channels=x_train.shape[3]             #input image is 28*28*1
input_pixels=input_width*input_height

n_conv1=32                                  #no. of units in conv layer 1
n_conv2=64

conv1_k=3                                   #filter size for conv layer 1
conv2_k=3

conv1_stride=1                              #stride for conv layer 1
conv2_stride=1

pool1_k=2                                   #pool size for pooling layer 1
pool2_k=2

n_hidden=1024                               #no. of units in hidden layer
n_output=10                                 #no. of units in output layer(there are 10 classes)

input_size_to_hiddenlayer=(input_height//(pool1_k*pool2_k))*(input_width//(pool1_k*pool2_k))*n_conv2
```

```python
def cnn_model():
    model=Sequential()
    
    #convolution layer 1:
    conv1=layers.Conv2D(filters=n_conv1,padding="same",
                        strides=[conv1_stride,conv1_stride],
                        activation="relu",use_bias=True,
                        kernel_size=[conv1_k,conv1_k],
                        input_shape=(input_height,input_width,input_channels))

    #convolution layer 2:
    conv2=layers.Conv2D(filters=n_conv1,
                        strides=[conv1_stride,conv1_stride],
                        activation="relu",use_bias=True,
                        kernel_size=[conv1_k,conv1_k],
                        )

    #max pooling layer 1:
    pool1=layers.MaxPooling2D(pool_size=[pool1_k,pool1_k],strides=[pool1_k,pool1_k],padding="same")
    
    #Dropout layer 1:
    dropout1=layers.Dropout(rate=0.25)
    
    #convolution layer 3:
    conv3=layers.Conv2D(filters=n_conv2,padding="same",
                        strides=[conv2_stride,conv2_stride],
                        activation="relu",use_bias=True,
                        kernel_size=[conv2_k,conv2_k],
                        )
    
    #convolution layer 4:
    conv4=layers.Conv2D(filters=n_conv2,padding="same",
                        strides=[conv2_stride,conv2_stride],
                        activation="relu",use_bias=True,
                        kernel_size=[conv2_k,conv2_k],
                        )

    #max pooling layer 2:
    pool2=layers.MaxPooling2D(pool_size=[pool2_k,pool2_k],strides=[pool2_k,pool2_k],padding="same")
    
    #Dropout layer 2:
    dropout2=layers.Dropout(rate=0.25)

    #flatten layer:                       
    flatten=layers.Flatten()

    #Hidden layer:
    hidden=layers.Dense(units=n_hidden,use_bias=True,activation="relu")

    #Dropout layer 3:
    dropout3=layers.Dropout(rate=0.5)

    #Output layer:
    output=layers.Dense(units=n_output,use_bias=True,activation="softmax")

    model.add(conv1)
    model.add(conv2)
    model.add(pool1)
    model.add(dropout1)
    model.add(conv3)
    model.add(conv4)
    model.add(pool2)
    model.add(dropout2)
    model.add(flatten)
    model.add(hidden)
    model.add(dropout3)
    model.add(output)
    
    model.summary()
    
    return model
```

```python
model=cnn_model()
```

# Compiling the Model:

```python
model.compile(optimizer=Adam(lr=0.001),loss="categorical_crossentropy",metrics=["accuracy"])
```

# Fit:

```python
model.fit(x_train,y_train,epochs=20,batch_size=128,validation_data=(x_test,y_test))
```

# Predict for test images:

```python
pred=model.predict(x_test)
pred
```

```python
score=model.evaluate(x_test,y_test)
score                                           #loss and accuracy for test data
```

# Classification Report:

```python
y_pred=np.argmax(pred,axis=1)
from collections import Counter
Counter(y_pred)
```

```python
#Confusion Matrix:
class_pred=np.array([class_names[i] for i in y_pred])     #converting numerical predictions into class names
class_true=np.array([class_names[i] for i in cls_test])     
skplt.metrics.plot_confusion_matrix(class_true,class_pred,figsize=(12,12))
```

```python
print(classification_report(cls_test,y_pred))
```
