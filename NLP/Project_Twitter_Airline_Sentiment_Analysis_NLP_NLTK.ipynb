```python
#removing inital "@"s. We can also let them be and not include the airline names as features.
x_train_text=[text.strip("@") for text in x_train_text]
x_test_text=[text.strip("@") for text in x_test_text]
x_train_text[0]
```

# Lemmatizing:

```python
#Getting the stopwords
from nltk.corpus import stopwords
import string
stop=stopwords.words("english")+list(string.punctuation)
stop
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet
lmtz=WordNetLemmatizer()

def getSimpleTag(tag):             
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("R"):
        return wordnet.ADV       
    else:
        return wordnet.NOUN
```

```python
def cleanWords(words):                                        #words is a list(of words in a particular document)
    clean_words=[]                                            #list of lemmatized,cleaned words to be returned
    for w in words:
        if w.isalpha() and w.lower() not in stop:
            pos=pos_tag([w])                                  #if w is provided as string rather than list,pos_tag will give tags for individual characters, instead of entire word
            simple_pos=getSimpleTag(pos[0][1])                #getting simpler tag
            lem_w=lmtz.lemmatize(w,simple_pos)                #lemmatized word
            clean_words.append(lem_w.lower())
    return clean_words
```

```python
x_train_text=[cleanWords(text) for text in x_train_text]
x_test_text=[cleanWords(text) for text in x_test_text]
print(x_train_text)
print()
print()
print(x_test_text)
```

# Selecting most occurring words as features (using CountVectorizer())

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=4000,ngram_range=(1,4),max_df=0.9)
```

```python
#converting (cleaned)word list back to string for each tweet
x_train_text=[" ".join(text_list) for text_list in x_train_text]
x_test_text=[" ".join(text_list) for text_list in x_test_text]
x_train_text                             
```
